### Join relevant tables to create a consolidated view.

```python
%%sql

SELECT b.book_ref, b.book_date, b.total_amount, t.passenger_id, tf.flight_id, f.flight_no, f.scheduled_departure
FROM ( bookings b
JOIN tickets t ON b.book_ref = t.book_ref
JOIN ticket_flights tf ON t.ticket_no = tf.ticket_no
JOIN flights f ON tf.flight_id = f.flight_id ) LIMIT 30

/* remove 'LIMIT 30' to see all rows */
```

## 15 SQL QUESTIONS

##### 1. Retrieve the total number of flights in the dataset.

```python
%%sql

SELECT COUNT(*) as total_flights FROM flights
```

##### 2. Find the average booking amount.

```python
%%sql

SELECT ROUND(AVG(total_amount),3) as avg_booking_amount FROM bookings
```

##### 3. List distinct aircraft codes.


```python
%%sql

SELECT DISTINCT aircraft_code FROM aircrafts_data
```

##### 4. Identify distinct airport codes.

```python
%%sql 

SELECT DISTINCT airport_code FROM airports_data LIMIT 25

/*Remove 'LIMIT 25' to see all rows of data*/
```

##### 5. Count the number of flights per status.

```python
%%sql

SELECT status, COUNT(*) as flight_count FROM flights GROUP BY status;

```

##### 6. Calculate the total booking amount per day.

```python
%%sql
SELECT DATE(book_date) as booking_day, SUM(total_amount) as total_amount FROM bookings GROUP BY booking_day;
```

##### 7. Retrieve flights with a specific aircraft code.

```python
%%sql

SELECT * FROM flights WHERE aircraft_code = 'CN1' LIMIT 20

/* Remove 'LIMIT 20' to see all rows of data */
```

##### 8. List the top 5 busiest airports based on departures.

```python
%%sql

SELECT departure_airport, COUNT(*) as departure_count FROM flights GROUP BY departure_airport ORDER BY departure_count DESC LIMIT 5
```

##### 9. Find flights that departed on time.

```python
%%sql

SELECT * FROM flights WHERE status = 'On Time' LIMIT 30

/* Remove 'LIMIT 30' to see all rows of data */
```

##### 10.  Identify the day with the highest number of cancellations

```python
%%sql

SELECT DATE(scheduled_departure) as day, COUNT(*) as cancellations FROM flights WHERE status = 'Cancelled' GROUP BY day ORDER BY cancellations DESC LIMIT 1
```

##### 11. Calculate the percentage of on-time arrivals per airport.

```python
%%sql

SELECT f.arrival_airport, ROUND((SUM(CASE WHEN f.actual_arrival <= f.scheduled_arrival THEN 1 ELSE 0 END) * 100.0 / COUNT(*)),2) AS on_time_percentage FROM flights f WHERE f.actual_arrival IS NOT NULL AND f.scheduled_arrival IS NOT NULL GROUP BY f.arrival_airport
```

##### 12. Find the longest flight (based on scheduled time).


```python
%%sql

SELECT f.flight_id, f.flight_no, f.scheduled_departure, f.scheduled_arrival, (f.scheduled_arrival - f.scheduled_departure) AS flight_duration FROM flights f ORDER BY flight_duration DESC LIMIT 1
```

##### 13. Determine the average flight range for each aircraft model

```python
%%sql

SELECT model, AVG(range) as avg_range FROM aircrafts_data GROUP BY model
```

##### 14. Calculate the total revenue generated by each aircraft model.

```python
%%sql

SELECT model, SUM(amount) as total_revenue FROM aircrafts_data ad JOIN flights f ON ad.aircraft_code = f.aircraft_code JOIN ticket_flights tf ON f.flight_id = tf.flight_id GROUP BY model
```

##### 15. Calculate the average delay time per airline.

```python
%%sql

SELECT f.aircraft_code, AVG(EXTRACT(EPOCH FROM (f.actual_arrival - f.scheduled_arrival))/60) AS avg_delay_minutes FROM flights f WHERE f.actual_arrival IS NOT NULL AND f.scheduled_arrival IS NOT NULL GROUP BY f.aircraft_code;
```
